# Using COVID-19 World Vaccination Progess Data to Learn Pandas

I have created this notebook for my own benefit to enhance my knowledge of pandas and its features as pertaining to data exploration.

# Importing the Data

```python
import pandas as pd 
covid = pd.read_csv('/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv')

#reviewing the first 5 rows of the data
covid.head()
```

# Exploring General Functions


**.melt(): unpivot a DF from wide to long format**

```python
# name = country, value = people vaccinated
covid.melt(id_vars=['country'], value_vars=['people_vaccinated'])
```

```python
# eg. unpivot multiple columns (date and people_vaccinated) as well as change column names
covid.melt(id_vars=['country'], value_vars=['date','people_vaccinated'], var_name='New_Var', value_name='New_Value')
```

**.pivot(): produces pivot tables based on 3 columns of the DF**

```python
covid.pivot('country','date','people_vaccinated')
```

**.crosstab(): Compute a simple cross tabulation of two (or more) factors. By default computes a frequency table of the factors unless an array of values and an aggregation function are passed.**

```python
# cross tabulation of the country and people_vaccinated columns from the covid data
pd.crosstab(covid.country, covid.people_vaccinated)
```

**.unique(): used to view all type of unique values in a certain column**

```python
# view all unique values of the country column
covid['country'].unique()
```

**pd.isna(): detect missing values for an array like object**

```python
# detect mising values within the covid table
pd.isna(covid)
```

**pd.to_numeric(): converts the whole database or sectioned database to a numeric value**

```python
# converting the people_vaccinated column from the covid table to a numeric value
pd.to_numeric(covid.people_vaccinated)
```

I won't be covering Series in this notebook as this mainly focuses on dataframes but I will state the difference between the two:
* Series can only contain a single list with an index
* A Dataframe can be made of more than one series (a collection of series)

```python
covid.index
```

**.dtypes: returns a series with the data type of each column**
* Columns with mixed types are stored with the 'object' data type

```python
# view the data types of each column within the covid dataframe
covid.dtypes
```

**.info(): prints a summary of a DF. including the index, column names, non null values, dtype and memory

```python
covid.info()
```

**.ndim: returns a number representing the number of array dimensions**
* Series has 1 dimension
* Dataframe has 2 dimensions

```python
covid.ndim
```

**.size: returns the number of elements in the array**

```python
covid.size
```

**.shape: returns a tuple representing the dimensionality of the DF**

```python
# there are 15 columns and 5236 rows in the covid dataframe
covid.shape
```

**.astype(): cast a pandas object to a specified dtype**

```python
# convert people_vaccinated to an object
object_vaccinated = covid['people_vaccinated'].astype(object)
object_vaccinated
```

**.at[]: access a single value for a row/column label pair**

```python
# access the value in the 5th row of the column labeled country
covid.at[5, 'country']
```

**.loc[]: access a group of rows and columns by label(s) or boolean array**

```python
covid.set_index('country', inplace=True)
covid
covid.loc['Canada']
```
